```python
# importing libraries :
import cv2
from matplotlib import pyplot
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Dense
from keras.layers import Flatten
from keras.optimizers import SGD
from keras.preprocessing.image import ImageDataGenerator
import os
import shutil 
import requests
import pandas as pd 
import csv 
import glob
import numpy as np
from keras.layers import Dropout
from keras.preprocessing.image import array_to_img , img_to_array , load_img
```

```python
import keras
from keras.callbacks import ModelCheckpoint , EarlyStopping , ReduceLROnPlateau
```

```python
# train path is the augmented images folder:
train_path="C:/Users/prince/train_images/"
test_path="C:/Users/prince/data/valid/"
```

# MODEL BUILDING (optimizer = SGD + Momentum)

```python
# MANUALLY RESIZING THE IMAGE SIZES FOR THE OUTPUT
i_wd=150
i_ht=150
# BUILDING CNN MODEL LAYERS
model = Sequential()

#FIRST CONV+RELU LAYER
model.add(Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(i_wd,i_ht,3)))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.25))


#SECOND CONV+RELU LAYER
model.add(Conv2D(64, (3, 3), activation='relu', padding='same', input_shape=(i_wd,i_ht,3)))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.25))

#THIRD CONV+RELU LAYER
model.add(Conv2D(64, (3, 3), activation='relu', padding='same', input_shape=(i_wd,i_ht,3)))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.25))


#FULLY CONNECTED LAYER
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(3, activation='softmax'))
```

```python
# CALLBACKS :

checkpoint= ModelCheckpoint(r"C:\Users\prince\data\check_sgd.h5",monitor="val_loss",mode="min",save_best_only=True,verbose=1)
earlystop= EarlyStopping(monitor="val_loss",min_delta=0,patience=3,verbose=1,restore_best_weights=True)
reduce_lr=ReduceLROnPlateau(monitor="val_loss",factor=0.2,patience=3,verbose=1,min_delta=0)

callbacks=[checkpoint,earlystop,reduce_lr]
```

```python
##  learning rate =0.001
```

```python
model.compile(loss="categorical_crossentropy",optimizer=SGD(momentum=True,learning_rate=0.001),metrics=['accuracy'])
```

```python
## DATA AUGMENTATION
train_data_gen=ImageDataGenerator(rescale=1./255)
valid_data_gen=ImageDataGenerator(rescale=1./255)
train_gen=train_data_gen.flow_from_directory(directory =train_path,target_size=(150, 150),batch_size=20,shuffle=True)
valid_gen=valid_data_gen.flow_from_directory(directory =test_path,target_size=(150, 150),batch_size=20,shuffle=True)
```

```python
#FITTING , GENERATING ACCURACY AND LOSS
history=model.fit(train_gen,validation_data=valid_gen,epochs = 20,callbacks=callbacks)
model.save("sgd1.h5")
```

```python
## score for the test data or validation data 
m=model.evaluate(valid_gen,verbose=0)
m
```

```python
# MANUALLY RESIZING THE IMAGE SIZES FOR THE OUTPUT
i_wd=150
i_ht=150
# BUILDING CNN MODEL LAYERS
model = Sequential()

#FIRST CONV+RELU LAYER
model.add(Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(i_wd,i_ht,3)))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.25))


#SECOND CONV+RELU LAYER
model.add(Conv2D(64, (3, 3), activation='relu', padding='same', input_shape=(i_wd,i_ht,3)))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.25))

#THIRD CONV+RELU LAYER
model.add(Conv2D(64, (3, 3), activation='relu', padding='same', input_shape=(i_wd,i_ht,3)))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.25))



#FULLY CONNECTED LAYER
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(3, activation='softmax'))
```

```python
checkpoint= ModelCheckpoint(r"C:\Users\prince\data\check_sgd.h5",monitor="val_loss",mode="min",save_best_only=True,verbose=1)
earlystop= EarlyStopping(monitor="val_loss",min_delta=0,patience=3,verbose=1,restore_best_weights=True)
reduce_lr=ReduceLROnPlateau(monitor="val_loss",factor=0.2,patience=3,verbose=1,min_delta=0)
callbacks=[checkpoint,earlystop,reduce_lr]

train_data_gen=ImageDataGenerator(rescale=1./255)
valid_data_gen=ImageDataGenerator(rescale=1./255)

train_gen=train_data_gen.flow_from_directory(directory =train_path,target_size=(150, 150),batch_size=20,shuffle=True)
valid_gen=valid_data_gen.flow_from_directory(directory =test_path,target_size=(150, 150),batch_size=20,shuffle=True)

## lr = 0.0001
model.compile(loss="categorical_crossentropy",optimizer=keras.optimizers.SGD(nesterov=True,learning_rate=0.0001),metrics=['accuracy'])


#FITTING , GENERATING ACCURACY AND LOSS
history=model.fit(train_gen,validation_data=valid_gen,epochs = 20,callbacks=callbacks)
model.save("nes1.h5")
```
