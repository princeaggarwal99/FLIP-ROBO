```python
# CREATING A FUNCTION FOR FINDING OUT THE BEST RANDOM STATE HAVING THE MAX r SCORE 
def max_score(classr,x,y):
    max_acc_score=0
    for r_state in range(42,101):
        x_train, x_test, y_train, y_test=train_test_split(x,y,random_state=r_state,test_size=0.30)
        classr.fit(x_train,y_train)
        y_pred=classr.predict(x_test)
        acc_scr=accuracy_score(y_test,y_pred)
        print('acc_score corr to random state',r_state,'is:',acc_scr)
        if acc_scr>max_acc_score:
            max_acc_score=acc_scr
            final_r_state=r_state
    print()
    print()
    print("max acc score corr to ",final_r_state,'is',max_acc_score)       
    return final_r_state 
```

```python
# FINDING  BEST RANDOM STATE FOR LINEAR REG WITH HIGH R SCORE
lreg=LogisticRegression()
r_state=max_score(lreg,x,y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,auc,roc_auc_score,roc_curve,f1_score
from sklearn.svm import SVC
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=63,test_size=.30)
```

```python
## tuning the parammeters
from sklearn.tree import DecisionTreeClassifier
rfr=RandomForestClassifier()
knn=KNeighborsClassifier(n_neighbors=10)
svc=SVC(kernel='rbf')
ada=AdaBoostClassifier()
gbr=GradientBoostingClassifier()
dtr=DecisionTreeClassifier()
lr=LogisticRegression()
```

```python
    models = []
models.append(('randomforestREG', rfr))
models.append(('kneighborsREG', knn))
models.append(('svc', svc))
models.append(('ada boost REG', ada))
models.append(('gradient boosting REG', gbr))
models.append(('decisiontree REG', dtr))
models.append(('logistic REG', lr))
```

```python
Model=[]
score=[]
cvs=[]
rocscore=[]
F1=[]
for name,model in models:
    print('********',name,'*******')
    print('\n')
    Model.append(name)
    model.fit(x_train,y_train)
    print(model)
    pre=model.predict(x_test)
    print('\n')
    AS=accuracy_score(y_test,pre)
    print("acc_score=",AS)
    score.append(AS*100)
    print('\n')
    f1=f1_score(y_test,pre)
    print("F1_score=",f1)
    F1.append(f1*100)
    print('\n')
    sc=cross_val_score(model,x,y,cv=10,scoring='accuracy').mean()
    print('cross_val_sc=',sc)
    cvs.append(sc*100)
    print('\n')
    fpr,tpr,thresholds=roc_curve(y_test,pre)
    roc_auc=auc(fpr,tpr)
    print("roc_auc_sc=",roc_auc)
    rocscore.append(roc_auc*100)
    print('\n')
    print('classification_repo\n',classification_report(y_test,pre))
    print('\n')
    cm=confusion_matrix(y_test,pre)
    print(cm)
    print('\n')
    plt.figure(figsize=(10,40))
    plt.subplot(911)
    plt.title(name)
    print(sns.heatmap(cm,annot=True))
    plt.subplot(912)
    plt.title(name)
    plt.plot(fpr,tpr)
    plt.plot([0,1],[0,1],'r--')
    plt.xlabel('TPR')
    plt.ylabel('FPR')
    print('\n\n')
```

```python
result=pd.DataFrame({'model':Model,'Accu_score':score,'f1_score':F1,'cross_val_Score':cvs,'roc-score':rocscore})
result
```
