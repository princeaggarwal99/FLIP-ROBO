# TASK-1

```python
# FETCHING  1 MONTH TWEETS BEFORE THE PANDEDMIC IN INDIA
def get_tweets_before_corona():
    import GetOldTweets3 as got
    tweetcri=got.manager.TweetCriteria().setQuerySearch('corona virus') \
    .setSince('2020-2-3') \
    .setUntil('2020-3-3') \
    .setMaxTweets(100)
    tweets=got.manager.TweetManager.getTweets(tweetcri)
    text_tweet=[[tweet.text] for tweet in tweets] 
    return text_tweet
```

```python
text=""
text_tweets=get_tweets_before_corona()
length=len(text_tweets)
for i in range(0,length):
    text=text_tweets[i][0] + " " + text
    
# converting to lowercase
lower_case = text.lower()

# Removing punctuations
cleaned_text = lower_case.translate(str.maketrans('', '', string.punctuation))

# splitting text into words

tokenized_words = cleaned_text.split()
```

```python
cleaned_text
```

```python
final_words= [word for word in tokenized_words if word not in stopwords]
final_words
```

```python
# fetching tweets after the pandemic occured in india of 1 month
def get_tweets_after_corona():
    import GetOldTweets3 as got
    tweetcri=got.manager.TweetCriteria().setQuerySearch('corona virus') \
    .setSince('2020-4-3') \
    .setUntil('2020-5-3') \
    .setMaxTweets(100)
    tweets=got.manager.TweetManager.getTweets(tweetcri)
    text_tweet=[[tweet.text] for tweet in tweets]
    return text_tweet
```

```python
text=""
text_tweets=get_tweets_after_corona()
length=len(text_tweets)
for i in range(0,length):
    text=text_tweets[i][0] + " " + text
    
# converting to lowercase
lower_case = text.lower()

# Removing punctuations
cleaned_text_2 = lower_case.translate(str.maketrans('', '', string.punctuation))

# splitting text into words
tokenized_words_2 = cleaned_text_2.split()
```

```python
cleaned_text_2 
```

```python
final_words2= [word for word in tokenized_words_2 if word not in stopwords]
final_words2
```

```python
## importing  emotions.txt file contanig emotions synonms for words
emotions=[]
with open('emotions.txt', 'r') as file:
    for line in file:
        clear_list=line.replace('\n','').replace(',','').replace("'",'').strip()
        word,emotion=clear_list.split(':')
        if word in final_words:
            emotions.append(emotion)
            
        
```

```python
# countering the emotions  before pandemic
from collections import Counter
w=Counter(emotions)
print(w)
```

```python
print("before pandemic sentiments ")
fig, ax1 = plt.subplots()
ax1.bar(w.keys(), w.values())
fig.autofmt_xdate()
plt.savefig('graph.png')
plt.show()
```
