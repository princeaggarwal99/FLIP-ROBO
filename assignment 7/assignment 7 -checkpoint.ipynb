```python
#importing libraries
import gensim
from gensim import corpora
from pprint import pprint
from gensim.models import Word2Vec
from gensim.utils import simple_preprocess
import os
import nltk
nltk.download('punkt')
import pprint
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# reading a document
data=open('stanford.txt',encoding='utf-8').read()
data
```

```python
# Preparing the dataset
# make it into a corpus by sent_Tokenize
corpus= nltk.sent_tokenize(data)
corpus
```

```python
stopwords=nltk.corpus.stopwords.words('english')
stopwords
```

```python
texts = [[word for word in document.lower().split() if word not in stopwords]
         for document in corpus]
pprint.pprint(texts)
```

```python
# making a dictionary :
dictionary = corpora.Dictionary(texts)


# Vectore representation 
pprint.pprint(dictionary.token2id)
```

# task-1 (similarity of atleast 20 tokens)

```python
# Training the Word2Vec model
model = Word2Vec(texts, min_count=1)


words = model.wv.vocab
```

```python
# taking some tokens from the corpus :
tokens=['problem','solving','climb','believed','offers','playing','higher','careers','respected','aware','innovation','bias','address','minority','diverse','social','communal','exit','study','insight','strategies','among','struggle','boost']
```

```python
## finding the similarities of these tokens :
for i in tokens:
    similar = model.wv.most_similar(i,topn=5)
    pprint.pprint(similar)
   
```

# task-2 : plotting 2-D graph to see words similarity of at least 20 words

```python
from sklearn.manifold import  TSNE
words =['problem','solving','climb','believed','offers','playing','higher','careers','respected','aware','innovation','bias','address','minority','diverse','social','communal','exit','study','insight','strategies','among','boost']
def tsne_plot(model):
    labels = []
    wordvecs = []

    for i in words:
        wordvecs.append(model[i])
        labels.append(i)
    
    tsne_model = TSNE(perplexity=3, n_components=2, init='pca', random_state=42)
    coordinates = tsne_model.fit_transform(wordvecs)

    x = []
    y = []
    for value in coordinates:
        x.append(value[0])
        y.append(value[1])
        
    plt.figure(figsize=(15,8)) 
    for i in range(len(x)):
        plt.scatter(x[i],y[i])
        plt.annotate(labels[i],
                     xy=(x[i], y[i]),
                     xytext=(2, 2),
                     textcoords='offset points', ha='left')

    plt.show()

tsne_plot(model)
```
